**Ülesanne 3: Insertion Sort praktikas (1p)<br>**
Antud on osaliselt sorteeritud loend: [12, 11, 13, 5, 6, 7]. Rakenda Insertion Sort algoritmi. <br>
Selgita, miks Insertion Sort võib olla tõhusam sellele loendile võrreldes täielikult sortimata loendiga. <br>
<br>
Insertion sort on adaptiivne algoritm, mis tähendab et juba osaliselt sorteeritud loendi puhul tõhusus tõuseb.

```python
def InsertionSort(arr):
    n = len(arr)  # loendi pikkus

    if n <= 1:
        return  # kui loendi pikkus on 1 või null

    for i in range(1, n):  # käib loendi läbi alates 2. elemendist

        key = arr[i]  # võti
        j = i-1       # võrdluseks eelnev element
        while j >= 0 and key < arr[j]:  # kui eelnev element ei ole esimene ja võti on väiksem kui eelnev element
            arr[j+1] = arr[j]  # vahetab asukoha
            j -= 1    # j = j-1
        arr[j+1] = key  # uus võti

arr =  [12, 11, 13, 5, 6, 7]
InsertionSort(arr)
print(arr)
```
